```python
def check(target):
    stack = []
    for ch in target:
        if ch == '(':
            stack.append(')')
        elif ch == '[':
            stack.append(']')
        elif ch == '{':
            stack.append('}')
        else:
            if len(stack) == 0:
                return False
            else:
                if stack.pop() != ch:
                    return False
    return len(stack) == 0
```

```python
def checkYear(year):
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    else:
        return year % 4 ==0

def check4(mmdd, year):
    mm = mmdd[:2]
    dd = mmdd[2:]
    if dd == '00':
        return False
    if mm in ('01', '03', '05', '07', '08', '10', '12'):
        return int(dd) <= 31
    elif mm in ('04', '06', '09', '11'):
        return int(dd) <= 30
    elif mm == '02':
        if int(dd) <= 28:
            return True
        elif dd == '29':
            return checkYear(year)
        else:
            return False
    else:
        return False;

def check3(mdd, year):
    mmdd = '0' + mdd
    return check4(mmdd, year)

def reverse(year):
    result = ''
    while year > 0:
        result = result  + str(year % 10)
        year = year / 10
    return result

def checkPalindrome(left, right):
    left = str(left)
    right = str(right)
    return left[0] == right[3] and left[1] == right[2] and left[2] == right[1] and (len(left)==3 or (left[3] == right[0]))


def countYearPalindrome(year):
    start = year / 100 * 100
    count = 0
    for i in range(100):
        current = start + i
        tmp = reverse(current)
        if check4(tmp, year) and checkPalindrome(tmp, current):
            print(tmp+str(current))
            count += 1
        if check3(tmp[:3], year) and checkPalindrome(tmp[:3], current):
            count += 1
    return count
    
```

```python
# Find GCC
g.components <- clusters(g)
print(g.components$csize)
# which is the largest component
ix <- which.max(g.components$csize)
# get the subgraph correspondent to just the giant component
g.giant <- induced.subgraph(g, which(g.components$membership == ix))
```
